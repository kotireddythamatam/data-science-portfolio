```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/bitcoin.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))

counts, bins, patches = plt.hist(
    df['open'],bins=10,
    color='violet',
    edgecolor='red',
    label='Open'
)

for i, count in enumerate(counts):
    plt.text((bins[i] + bins[i+1]) / 2, count, str(count), ha="center", va='bottom')

plt.title('Bitcoin Open Distibution')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

counts, bins, patches = plt.hist(
    df['high'],
    label='High'
)

for i,count in enumerate(counts):
    plt.text((bins[i] + bins[i+1]) / 2, count, str(count), ha='center', va='top')

plt.title('Bitcoin High Distibution')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(
    df['open'],
    df['high']
)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(
    df['open'],
    df['low']
)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(
    df['open'],
    df['close']
)

plt.xlabel('Open')
plt.ylabel('Close')


plt.show()
```
