```python
import pandas as pd
```

```python
df = pd.read_csv("../../datasets/Bengaluru_House_Data.csv")
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Previous value will be filled
df["location"].fillna('ffill', inplace=True)
```

```python
# after value will be filled
df["size"].fillna('bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
# filling null values for different columns with different values
df.fillna({"bath": df["bath"].mean(), "balcony": df["balcony"].median()}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(how="all", inplace=True)
```
