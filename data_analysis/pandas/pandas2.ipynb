```python
import pandas as pd
```

```python
# Series without index access, default 0, 1 etc
data = [1,2,3,4,5]
df = pd.Series(data)
df[0]
```

```python
# Series index access
data = [1,2,3,4,5,6]
df = pd.Series(data, index=['a','b','c','d','e','f'])
df['a']
```

```python
# Series with dict, keys will become indexes
data = {"a": 1, "b":2, "c":3, "d":4}
df = pd.Series(data)
df
```

```python
# Series we can mention some items based on index
data = {"a": 1, "b":2, "c":3, "d":4}
df = pd.Series(data, index=["a", "b"])
df
```

```python
# Dataframe with dict, keys will become lables
data = {
    "a": [1,2,3,4,5],
    "b": [11,22,33,44,55]
}
df = pd.DataFrame(data)
df
```

```python
# Dataframe with loc to get record
data = {
    "a": [1,2,3,4,5],
    "b": [11,22,33,44,55]
}
df = pd.DataFrame(data, index = ["i1", "i2", "i3", "i4", "i5"])
#df.loc[0]
# df.loc[[0,2]]
df.loc['i4']
```
