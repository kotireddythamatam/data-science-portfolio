# Convolutional Neural Network (CNN) to classify CIFAR images

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import datasets
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# Split X and y
(X_train, X_test), (y_train, y_test) = datasets.cifar10.load_data()
```

```python
# Normalize pixel values to be between 0 and 1
X_train = X_train / 255
y_train = y_train / 255
```

```python
# Create Convolution base
model = Sequential([
    InputLayer(shape=(32, 32, 3)),
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    Flatten(),
    Dense(units=64, activation='relu'),
    Dense(units=10)
])
```

```python
model.summary()
```

```python
# Compile model
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
# Train model
model_history = model.fit(X_train, X_test, epochs=10)
```

```python
test_loss, test_accuracy = model.evaluate(y_train, y_test, verbose=2)
```

```python
plt.plot(model_history.history['accuracy'], label='accuracy')
plt.plot(model_history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
```
