```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Load dataset
data = {
    'Age': [22, 25, 47, 52, 46, 56, 55, 60],
    'Salary': [25000, 30000, 47000, 52000, 48000, 58000, 60000, 62000],
    'Buy': [0, 0, 1, 1, 1, 1, 0, 0]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Split independant(X) and dependant(y) sets
X = df.iloc[:, 0:2]
y = df.iloc[:, 2]
```

```python
# Train–Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Scale X (only X, not y!)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Build ANN model
model = Sequential([
    Dense(units=8, activation='relu', input_shape=(2,)),
    Dense(units=1, activation='sigmoid')
])
```

```python
# Compile Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train (Fit the model)
model_history = model.fit(X_train, y_train, epochs=100, batch_size=10)
```

```python
# Predict on Test Data
y_pred = model.predict(X_test)
y_pred = (y_pred>=0.5).astype(int)
y_pred
```

```python
# Evaluate Model
loss, accuracy = model.evaluate(X_test, y_test)
```
