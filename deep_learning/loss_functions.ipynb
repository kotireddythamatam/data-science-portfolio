# Loss Funtions

```python
# Import libraries
import math
```

```python
def mse_loss_function(y_true, y_predicted):
    return (y_true - y_predicted)**2
```

```python
mse_loss_function(4, 3.5)
```

```python
mse_loss_function(2, 5)
```

```python
def mae_loss_function(y_true, y_predicted):
    return abs(y_true - y_predicted)
```

```python
mae_loss_function(4, 3.5)
```

```python
mae_loss_function(2, 5)
```

```python
def bce_loss_function(y_true, y_predicted):
    return -(y_true * math.log(y_predicted) + (1 - y_true) * math.log(1 - y_predicted))
```

```python
bce_loss_function(1, 0.8)
```

```python
def cce_loss_function(y_true, y_predicted):
    return -sum(y_true[i] * math.log(y_predicted[i]) for i in range(len(y_true)))
```

```python
cce_loss_function([0, 0, 1], [0.1, 0.2, 0.7])
```
