# Forward Propagation and back propagation

```python
# Import libraries
import time

# Training data (x, y)
x = 2
y = 5   # because true relation is y = 2*x + 1 -> 2*2 + 1 = 5

def train_2x_plus_1_model(w, b, lr=0.1):
    print(f"Initial w: {w:.4f}, initial b: {b:.4f}, initial lr: {lr:.4f}\n")
    epoch = 0
    while True:
        
        # ------ Forward Propagation ---------
        fp = w * x + b
        loss = (y - fp) ** 2
    
        # ------ Back Propagation -----------
        dL_dw = 2 * (y - fp) * (-x)
        dL_db = 2 * (y - fp) * (-1)
    
        # ------ Update weight and bias -----
        w = w - lr * dL_dw
        b = b - lr * dL_db
    
        # ---- Print progress ----
        epoch += 1
        print(f"Epoch: {epoch}")
        print(f"Predicted: {fp}, Loss: {loss}")
        print(f"Updated w: {w}, Updated b: {b}\n")
        if loss == 0:
            break
```

```python
train_2x_plus_1_model(w=2, b=1, lr=0.1)
```

```python
#training_data(x1, x2, y)
x1 = 1
x2 = 0
y = 3        # because y = 2 * 1 + 3 * 0 + 1 = 2 + 0 + 1 = 3

def train_2x_plus_3x_plus_1_model(w1, w2, b, lr=0.1):
    print(f"initial w1: {w1}, w2: {w2}, b: {b}, lr: {lr}")
    iteration = 0
    while True:

        # ----- Forward Propagation ----
        fp = w1 * x1 + w2 * x2 +b
        loss = (y - fp) ** 2

        # ----- Back Propagation -----
        dL_dw1 = 2 * (y - fp) * (-x1)
        dL_dw2 = 2 * (y - fp) * (-x2)
        dL_db = 2 * (y - fp) * (-1)

        # ---- Update weights and bias ----
        w1 = w1 - lr * dL_dw1
        w2 = w2 - lr * dL_dw2
        b = b - lr * dL_db

        # print new updates
        print(f"iteration: {iteration}")
        print(f"prediction: {fp}, loss: {loss}")
        print(f"updated w1: {w1}, w2: {w2}, b: {b}")
        iteration += 1
        if loss == 0:
            break
        
```

```python
train_2x_plus_3x_plus_1_model(w1=2, w2=3, b=1, lr=0.01)
```

```python
# training data(x, y)
X = [1, 2, 3, 4]
Y = [3, 5, 7, 9]

def train_data_set_2x_plus_one_model(w, b, lr=0.1):
    print(f"initial w: {w}, b: {b}, lr: {lr}")
    iteration = 0
    while True:
        for x, y in zip(X, Y):

            # ----- Forward Propagation -----
            fp = w * x + b
            loss = (y - fp) ** 2

            #----- Back Propagation -----
            dw = 2 * (y - fp) * (-x)
            db = 2 * (y - fp) * (-1)

            # ----- Update weights and bias -----
            w = w - lr * dw
            b = b - lr * db

            # print data
            print(f"iteration: {iteration}")
            print(f"prediction: {fp}, loss: {loss}")
            print(f"updated w: {w}, b: {b}")
            iteration += 1
        if loss == 0:
            break
```

```python
train_data_set_2x_plus_one_model(w=1, b=1, lr=0.1)
```
